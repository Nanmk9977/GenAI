 # 📌 Mastering Prompt Templates in LangChain

This project demonstrates the development of a **Mini Prompt Engine** using LangChain, focusing on building dynamic and reusable prompt systems. The implementation replaces prompts with  templates using `PromptTemplate` and `ChatPromptTemplate`.

**1.Install & Imports**

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
from langchain_core.prompts import PromptTemplate

test_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(test_template.format(topic="AI"))
```

**Task 1 – Replace Hardcoded Prompt**

```python
# Reusable template (NO f-strings)
basic_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

def explain_topic(topic):
    return basic_template.format(topic=topic)

# Test
print(explain_topic("Artificial Intelligence"))
```

**Task 2 – Multi-Input Prompt System**

```python
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

def generate_multi_prompt(topic, audience, tone):
    return multi_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test Cases
print(generate_multi_prompt("AI", "beginners", "friendly"))
print(generate_multi_prompt("Python", "kids", "fun"))
print(generate_multi_prompt("Deep Learning", "engineers", "technical"))
```

**Task 3 – Prompt Variations Engine**

```python
# Define templates
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

def get_prompt_variations(topic):
    return {
        "Teaching": teaching_template.format(topic=topic),
        "Interview": interview_template.format(topic=topic),
        "Storytelling": story_template.format(topic=topic)
    }

# Test
result = get_prompt_variations("Machine Learning")
for key, value in result.items():
    print(f"{key}: {value}")
```

**Task 4 – ChatPromptTemplate System**

```python
def build_chat_prompt(role, topic):

    role_map = {
        "teacher": "You are a helpful and patient teacher.",
        "interviewer": "You are a professional interviewer.",
        "motivator": "You are an inspiring motivational speaker."
    }

    system_message = role_map.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# Test
messages = build_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

**Task 5 – Input Validation Layer**

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Normalize input (avoid case errors)
    audience = audience.lower()
    tone = tone.lower()

    if audience not in valid_audience:
        raise ValueError(f"Invalid audience: {audience}")

    if tone not in valid_tone:
        raise ValueError(f"Invalid tone: {tone}")

    return audience, tone

# Test
try:
    validate_inputs("Beginner", "Fun")
    print("Inputs are valid ✅")
except Exception as e:
    print(e)
```

**Task 6 – Prompt Generator App**

```python
def generate_prompt(topic, audience, tone, style):

    # Validate first
    audience, tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style."
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} tone."
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style."
        )
    }

    if style not in style_templates:
        raise ValueError("Style must be: teaching / interview / storytelling")

    return style_templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7 – Template Reusability Test**

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}."
)

test_inputs = [
    ("AI", "beginners"),
    ("Python", "students"),
    ("Data Science", "analysts"),
    ("Machine Learning", "engineers"),
    ("Deep Learning", "researchers")
]

for topic, audience in test_inputs:
    print(reusable_template.format(topic=topic, audience=audience))
```

**Final Pipeline**

```python
def prompt_pipeline(topic, audience, tone, style):

    # Step 1: Validate
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Generate Prompt
    final_prompt = generate_prompt(topic, audience, tone, style)

    return final_prompt

# Test Pipeline
print(prompt_pipeline("Artificial Intelligence", "beginner", "fun", "teaching"))
```
